```python
import matplotlib.pyplot as plt 
import numpy as np
```

# 1.  Draw a Line Plot 

```python
x = [1,2,3,9,10]
y = [1,7,9,8,5]
plt.plot(x,y)
plt.show()
```

# 2. Give Title, Xlabel and Ylabel

```python
x = [1,2,3,9,10]
y = [1,7,9,8,5]
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

# 3. Change Font Size of Title, Xlabel and Ylabel and also change location.

```python
x = [1,2,3,9,10]
y = [1,7,9,8,5]
plt.plot(x,y)
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.show()
```

# 4. Change color & width of the line

```python
x = [1,2,3,9,10]
y = [1,7,9,8,5]
plt.plot(x,y,color='r',linewidth=4)
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.show()
```

# 5. Change Marker size and line style.

**Markers**

=============   ===============================
character       description
=============   ===============================
``'.'``         point marker
``','``         pixel marker
``'o'``         circle marker
``'v'``         triangle_down marker
``'^'``         triangle_up marker
``'<'``         triangle_left marker
``'>'``         triangle_right marker
``'1'``         tri_down marker
``'2'``         tri_up marker
``'3'``         tri_left marker
``'4'``         tri_right marker
``'8'``         octagon marker
``'s'``         square marker
``'p'``         pentagon marker
``'P'``         plus (filled) marker
``'*'``         star marker
``'h'``         hexagon1 marker
``'H'``         hexagon2 marker
``'+'``         plus marker
``'x'``         x marker
``'X'``         x (filled) marker
``'D'``         diamond marker
``'d'``         thin_diamond marker
``'|'``         vline marker
``'_'``         hline marker
=============   ===============================

**Line Styles**

=============    ===============================
character        description
=============    ===============================
``'-'``          solid line style
``'--'``         dashed line style
``'-.'``         dash-dot line style
``':'``          dotted line style
=============    ===============================


```python
x = [1,2,3,9,10]
y = [1,7,9,8,5]
plt.plot(x,y,color='r',linewidth=4,marker='>',markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',
        linestyle=':')
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.show()
```

```python
# We can also combine color, marker, linestyle parameter like this as shown below
x = [1,2,3,9,10]
y = [1,7,9,8,5]    # we can see this after x,y all parameters are used together
plt.plot(x,y,'r--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue')
       
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.show()
```

# 6. Add legend to your graph also change location


```python
x=np.random.randint(low=1, high=10, size=20)
y=np.random.randint(low=1, high=8, size=20)
plt.plot(x,y,'r--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue')
       
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.legend(["A simple line plot"],loc=2)
plt.show()
```

# 8. Xlim and Ylim

```python
x=np.random.randint(low=1, high=10, size=20)
y=np.random.randint(low=1, high=8, size=20)
plt.plot(x,'r--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 2")     
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xlim(0,20)
plt.ylim(0,15)
plt.legend()
plt.show()
```

# 9. Change X and Y limit Using axis Method

```python
x=np.random.randint(low=1, high=10, size=20)
y=np.random.randint(low=1, high=8, size=20)
plt.plot(x,'r--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 2")     
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
# xmin, xmax, ymin, ymax
plt.axis([0,20,0,15])
plt.legend()
plt.show()
```

# 10. Change X and Y limit Using Xticks and Yticks

```python
x_ticks=[x for x in range (0,20,2)]
y_ticks=[y for y in range (0,15,3)]
```

```python
x=np.random.randint(low=1, high=10, size=20)
y=np.random.randint(low=1, high=8, size=20)
plt.plot(x,'r--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 2")     
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.show()
```

# 11. Use Grid Method

```python
x=np.random.randint(low=1, high=10, size=20)
y=np.random.randint(low=1, high=8, size=20)
plt.plot(x,'r--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 2")     
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color='b',linestyle='--',linewidth=1)
plt.show()
```

# 12. Change Figure size


```python
x=np.random.randint(low=1, high=10, size=20)
y=np.random.randint(low=1, high=8, size=20)
plt.figure(figsize=(10,3),facecolor='lightgreen',edgecolor='black',
          linewidth=1)
plt.plot(x,'r--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 1")
plt.plot(y,'b--<',linewidth=4,markersize=10,
        markerfacecolor='blue',markeredgecolor='blue',label="Line 2")     
plt.title("Line Plot",fontsize=20,loc='left')
plt.xlabel("X-Axis",fontsize=20,loc='right')
plt.ylabel("Y-Axis",fontsize=20,loc='bottom')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color='b',linestyle='--',linewidth=1)
plt.show()
```
